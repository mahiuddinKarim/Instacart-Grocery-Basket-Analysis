```python
# task 4 creating a loyalty flag for existing customers using the transform() and loc() functions.
```

```python
#Transforming 

ords_prods_merge['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# data set cheaking
ords_prods_merge.head(50)
```

```python

#Creating and assigning the loyalty flag in the new column
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking frequency of flag

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
#Checking data frame
ords_prods_merge.head(50)
```

```python
# Task 5  loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer)
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

```python
# task 6 put flag as low spender and high spender for  marketing
```

```python
# Generating summary stats (used .head due to high number of customers)

ords_prods_merge.groupby('user_id').agg({'prices' : ['mean', 'min', 'max']}).head(20)
```

```python
# Transforming
ords_prods_merge['spending'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Checking data frame
ords_prods_merge.head(50)
```

```python
# Creating flags as 'low spender' and 'high spender'

ords_prods_merge.loc[ords_prods_merge['spending'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['spending'] > 10, 'spending_flag'] = 'High spender'
```

```python
# cheaking data frame
ords_prods_merge.head(50)
```

```python
# task 7 Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column
```

```python

ords_prods_merge.groupby('user_id').agg({'days_since_prior_order' : ['mean']})
```

```python
#transformation 
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.max)
```

```python
# cheaking data frame
ords_prods_merge.head(50)
```

```python
# Creating Flag  as 'Non-frequent customer',
ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] < 20, 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```
